```python
#API DOCUMENTATION: https://www.visualcrossing.com/resources/documentation/weather-data/weather-data-documentation/
import pandas as pd
import requests
import time

# Set your Visual Crossing Weather API key
api_key = "NT5LXX7GEVMWMTYD6H4DKAE57"

# Set input and output CSV file paths
input_csv = "/content/finalzipcodes-hawaii.csv"
output_csv = "output_annual_weather_data.csv"

# Read the input CSV file containing zipcodes
input_data = pd.read_csv(input_csv)

# Set up the output dataframe
output_data = pd.DataFrame(columns=['Zipcode', 'Average Annual Max Temp', 'Average Annual Min Temp', 'Average Annual Precipitation', 'Average Annual Wind Speed', 'Average Annual UV Index', 'Average Annual Visibility', 'Average Annual Cloud Cover', 'Average Annual Humidity', 'Average Annual Solar Radiation'])

# Function to fetch weather data
def fetch_weather_data(zipcode, api_key, year):
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{zipcode}/{year}-01-01/{year}-12-31?unitGroup=us&key={api_key}&include=obs"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching weather data for zipcode {zipcode}: {response.status_code} {response.text}")
        return None

# Function to calculate average annual values
def get_average_annual_values(weather_data):
    days = weather_data['days']
    max_temps = [day['tempmax'] for day in days if day['tempmax'] is not None]
    min_temps = [day['tempmin'] for day in days if day['tempmin'] is not None]
    precipitations = [day['precip'] for day in days if day['precip'] is not None]
    wind_speeds = [day['windspeed'] for day in days if day['windspeed'] is not None]
    uv_indexes = [day['uvindex'] for day in days if day['uvindex'] is not None]
    visibilities = [day['visibility'] for day in days if day['visibility'] is not None]
    cloud_covers = [day['cloudcover'] for day in days if day['cloudcover'] is not None]
    humidities = [day['humidity'] for day in days if day['humidity'] is not None]
    solar_radiations = [day['solarradiation'] for day in days if day['solarradiation'] is not None]

    max_temp_avg = sum(max_temps) / len(max_temps) if max_temps else None
    min_temp_avg = sum(min_temps) / len(min_temps) if min_temps else None
    precipitation_avg = sum(precipitations) / len(precipitations) if precipitations else None
    wind_speed_avg = sum(wind_speeds) / len(wind_speeds) if wind_speeds else None
    uv_index_avg = sum(uv_indexes) / len(uv_indexes) if uv_indexes else None
    visibility_avg = sum(visibilities) / len(visibilities) if visibilities else None
    cloud_cover_avg = sum(cloud_covers) / len(cloud_covers) if cloud_covers else None
    humidity_avg = sum(humidities) / len(humidities) if humidities else None
    solar_radiation_avg = sum(solar_radiations) / len(solar_radiations) if solar_radiations else None

    return max_temp_avg, min_temp_avg, precipitation_avg, wind_speed_avg, uv_index_avg, visibility_avg, cloud_cover_avg, humidity_avg, solar_radiation_avg  

# Set the year for which to fetch the weather data
year = "2022" 

#Iterate through each zipcode and fetch weather data
for index, row in input_data.iterrows():
    zipcode = row['Zipcode']
    weather_data = fetch_weather_data(zipcode, api_key, year)

    if weather_data:
        max_temp_avg, min_temp_avg, precipitation_avg, wind_speed_avg, uv_index_avg, visibility_avg, cloud_cover_avg, humidity_avg, solar_radiation_avg = get_average_annual_values(weather_data)
        output_data = output_data.append({'Zipcode': zipcode, 'Average Annual Max Temp': max_temp_avg, 'Average Annual Min Temp': min_temp_avg, 'Average Annual Precipitation': precipitation_avg, 'Average Annual Wind Speed': wind_speed_avg,'Average Annual UV Index': uv_index_avg,'Average Annual Visibility': visibility_avg, 'Average Annual Cloud Cover': cloud_cover_avg, 'Average Annual Humidity': humidity_avg, 'Average Annual Solar Radiation': solar_radiation_avg }, ignore_index=True)

    time.sleep(1)  # Adding a delay to avoid rate limiting

# Save the output dataframe to a CSV file
output_data.to_csv(output_csv, index=False)
```
